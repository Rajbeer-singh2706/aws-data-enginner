```python
import os
from pyspark.sql import SparkSession

# ---------------------------
# Windows Hadoop setup
# ---------------------------
os.environ["HADOOP_HOME"] = r"C:\hadoop"
os.environ["PATH"] += os.pathsep + r"C:\hadoop\bin"

# ---------------------------
# Spark Session
# ---------------------------
spark = (
    SparkSession.builder
    .appName("CSV_to_Parquet_Windows")
    .master("local[*]")
    .config("spark.sql.adaptive.enabled", "false")
    .config("spark.sql.shuffle.partitions", "8")
    .config("spark.hadoop.fs.file.impl",
            "org.apache.hadoop.fs.LocalFileSystem")
    .config("spark.hadoop.fs.raw.impl",
            "org.apache.hadoop.fs.RawLocalFileSystem")
    .config("spark.hadoop.fs.permissions.umask-mode", "000")
    .config("dfs.permissions.enabled", "false")
    .getOrCreate()
)

# ---------------------------
# Read CSV
# ---------------------------
csv_path = r"E:\data\orders_large_dataset\orders_1.csv"

df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(csv_path)
)

#print("CSV count:", df.count())
df.show(5)

# ---------------------------
# Write Parquet with coalesce
# ---------------------------
parquet_path = r"E:\data_new\orders_large_dataset\parquet_output"

(
    df
    .write
    .mode("overwrite")
    .parquet(parquet_path)
)

print("Parquet write successful ✅")

spark.stop()
```
